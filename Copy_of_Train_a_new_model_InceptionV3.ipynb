```
# <------ Compile this cell clicking here

#Now that you have compiled previous cells, we can train the model

###### CELL TRAINING ######

# Don't close the web page during training !


# Once this cell is finished compiling, your trained model and its labels files will be available as Model_trained.tflite + Model_trained.txt in the "Files" left panel
# Just download them and import them on your raspberrypi

# This install some librairies for deep Learning. It should takes maximum one minute
!pip install tflite-model-maker


NameModel="Model_Glue1_0"
NumberOfEpochs=10
BatchSize=32


# from tflite_model_maker import configs
# from tflite_model_maker import ExportFormat
from tflite_model_maker import image_classifier
from tflite_model_maker import ImageClassifierDataLoader
from tflite_model_maker import model_spec

import matplotlib.pyplot as plt

inception_v3_spec = model_spec.ImageModelSpec(uri='https://tfhub.dev/google/imagenet/inception_v3/feature_vector/1')
inception_v3_spec.input_image_shape = [299, 299]

Tensorflow_Model_to_FineTune=inception_v3_spec #efficientnet_lite1

data = ImageClassifierDataLoader.from_folder(image_path)
train_data, test_data = data.split(0.9)


model = image_classifier.create(train_data,epochs=NumberOfEpochs,batch_size=BatchSize,model_spec=Tensorflow_Model_to_FineTune,dropout_rate=0.2)
loss, accuracy = model.evaluate(test_data,batch_size=BatchSize)

model.export(export_dir='Model_Trained',tflite_filename=str(NameModel+".tflite"), label_filename=str(NameModel+".txt"),with_metadata=False)


## End of script. Your model trained will be available on the left pannel in "Files"
```
